```python


alphabet = ['(', ')', 'a', 'b', '*', '|']

def bases(): ## return a list of base chars, Level 0
    ## return ['a', 'b']
    ## Initially we thought of having multiple base charaters but later
    ## we decided to have only one base character 'a' because we can still 
    ## generate all the structures of regexes with only one base character
    ## and then we can plug some ranges or sets like [2-3] and [035] randomly.
    ## e.g. ((a)*)|((a*)*) --> (([2-3])*)|(([035]*)*)

    return ['a']

l0 = bases()
print(len(l0))
l0
```

```python

## Level 1 regexes building on top of Level 0 regexes
def rgx1(r1):  
    
    r2 = [] + r1
    x = r1[0]
    r2.append(f'{x}*') ## Kleene star
    r2.append(f'{x}+') ## Kleene plus
    r2.append(f'{x}?') ## Optional
    for i,r in enumerate(r1): 
        for s in (r1[i:]): 
            if r == s and r != x: continue 
            r2.append(f'{r}|{s}') ## Alternation
            # r2.append(f'{r}{s}') ## Concatenation is not counted in this level 
            ## as 'aa' and 'a' are the same regexes in this level
    return r2

# l1 = list(set(rgx1(l0))) 
l1 = rgx1(l0)
print(len(l1))
l1
```

```python

## Level 2 regexes building on top of Level 1 regexes
def rgx2(r1): 
    r2 = [] + r1
    for i,r in enumerate(r1):
        if len(r) != 1:  
            r2.append(f'({r})*')
            r2.append(f'({r})+')
            r2.append(f'({r})?')     
        for s in (r1[i:]): 
            if r == s and r == 'a': continue
            r2.append(f'({r})|({s})') 
            r2.append(f'({r})({s})') ## Concatenation
    
    return r2

l2 = rgx2(l1)
print(len(l2))
l2

# count = 0
# for r in l2: 
#     if r.count('*') == 1 and r.count('|') == 1: 
#         count += 1
#         print(r)

# print(count)
```

```python
def rgx3(r1): 
    
    r2 = [] + r1
    for i,r in enumerate(r1): 
        r2.append(f'({r})*')
        r2.append(f'({r})+')
        r2.append(f'({r})?')     
        for s in (r1[i:]): 
            # if r == s : continue
            r2.append(f'({r})|({s})') 
            r2.append(f'({r})({s})')
    
    return r2

l3 = rgx3(l2)
print(len(l3))
l3

# count = 0
# for r in l3: 
#     if r.count('|') == 3 and r.count('*') == 0: 
#         count += 1
#         print(r)

# print(count)
```

```python
def rgx4(r1): 
    
    r2 = [] + r1
    for i,r in enumerate(r1): 
        r2.append(f'({r})*')      
        for s in (r1[i:]): 
            r2.append(f'({r})|({s})') 
            r2.append(f'({r})({s})')
    
    return r2

l4 = rgx4(l3)
print(len(l4))
```
